```python
conn = sql.connect("ad65_voters.db")
```

```python
df.to_sql("voters", conn, if_exists='replace', index=False)
```

```python
# Youth voters (born after 2000) with email addresses
youth_select = pd.read_sql_query("""
select * from voters 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- proxy for born after 2000 (under 25)
;
""", conn)
youth_select
```

```python
youth_select.to_csv('ad65_youth_with_email.csv', index=False, encoding='utf-8')
```

```python
# Youth who voted recently (have a LastElectionVotedIn value)
youth_recent = pd.read_sql_query("""
select * from voters 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- proxy for born after 2000 (under 25)
and LastElectionVotedIn is not NULL
;
""", conn)
youth_recent
```

```python
youth_recent.to_csv('ad65_youth_recent_voters.csv', index=False, encoding='utf-8')
```

```python
# Youth who haven't voted (no LastElectionVotedIn)
youth_no_vote = pd.read_sql_query("""
select * from voters 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- proxy for born after 2000 (under 25)
and LastElectionVotedIn is NULL
;
""", conn)
youth_no_vote
```

```python
youth_no_vote.to_csv('ad65_youth_no_vote_history.csv', index=False, encoding='utf-8')
```
