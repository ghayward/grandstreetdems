```python
conn = sql.connect("gsd_dems_1.db")
```

```python
dems.to_sql("gsdems",conn,if_exists='replace',index=False)
```

```python
gsd_dems_youth_select_1 = pd.read_sql_query("""
select * from gsdems 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- this is our proxy for born after 2000 (so under 25)
and ElectionDistrict in (013,014,023,024,025,026,028,029,030,031,032,033,041,046,047,048,052,053,054,067)
;
""", conn)
gsd_dems_youth_select_1
```

```python
gsd_dems_youth_select_1.to_csv('gsd_dems_youth_select_1.csv', index=False, encoding='utf-8')
```

```python
gsd_dems_youth_activist = pd.read_sql_query("""
select * from gsdems 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- this is our proxy for born after 2000 (so under 25)
and ElectionDistrict in (013,014,023,024,025,026,028,029,030,031,032,033,041,046,047,048,052,053,054,067)
and PR20240625 != 'N' --PR20240625 means 2024 Primary; 'N' means didn't vote
;
""", conn)
gsd_dems_youth_activist
```

```python
gsd_dems_youth_activist.to_csv('gsd_dems_youth_activist.csv', index=False, encoding='utf-8')
```

```python
gsd_dems_youth_missed_2024_primary = pd.read_sql_query("""
select * from gsdems 
where 1=1
and PoliticalParty in ('DEM', 'WOR')
and EmailAddress is not NULL
and BirthDate >= 20000000 -- this is our proxy for born after 2000 (so under 25)
and ElectionDistrict in (013,014,023,024,025,026,028,029,030,031,032,033,041,046,047,048,052,053,054,067)
and PR20240625 = 'N' --PR20240625 means 2024 Primary; 'N' means didn't vote
;
""", conn)
gsd_dems_youth_missed_2024_primary
```

```python
gsd_dems_youth_missed_2024_primary.to_csv('gsd_dems_youth_missed_2024_primary.csv', index=False, encoding='utf-8')
```
